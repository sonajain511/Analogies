```python
# random_samples = random.sample(range(0,len(file_contents)), 30)
random_samples = [38, 332, 365, 301, 279, 64, 102, 114, 431, 320, 384, 70, 178, 161, 77, 172, 7, 269, 85, 94, 288, 399, 41, 177, 318, 240, 140, 54, 201, 73]
print(random_samples)
ratio_list = []
index_list = []
data = []
```

```python
for index in range(len(random_samples)):
  a_glove_rep = model.encode(a[random_samples[index]].rstrip())
  b_glove_rep = model.encode(b[random_samples[index]].rstrip())
  c_glove_rep = model.encode(c[random_samples[index]].rstrip())
  d_glove_rep = model.encode(d[random_samples[index]].rstrip())

  a_b_cosine = 1 - scipy.spatial.distance.cosine(a_glove_rep, b_glove_rep)
  c_d_cosine = 1 - scipy.spatial.distance.cosine(c_glove_rep, d_glove_rep)

  a_c_cosine = 1 - scipy.spatial.distance.cosine(a_glove_rep, c_glove_rep)
  b_d_cosine = 1 - scipy.spatial.distance.cosine(b_glove_rep, d_glove_rep)

  min_a_b_c_d = min(a_b_cosine/c_d_cosine, c_d_cosine/a_b_cosine)
  min_a_c_b_d = min(a_c_cosine/b_d_cosine, b_d_cosine/a_c_cosine)

  # ratio_list.append(min(a_b_cosine, c_d_cosine))
  # index_list.append(index + 1)

  analogy = a[random_samples[index]] + ":" + b[random_samples[index]] + "::" + c[random_samples[index]] + ":" + d[random_samples[index]]
  # data.append([abs(min(a_b_cosine/c_d_cosine, c_d_cosine/a_b_cosine)), my_response[index], analogy])
  data.append([max(min_a_b_c_d, min_a_c_b_d), my_response[index], analogy])

print(analogy)
```

```python
df = pd.DataFrame(data, columns = ['Ratio', 'Response', "Analogy"])
fig = px.scatter(df, x = "Analogy", y = "Ratio")
fig.show()
```
